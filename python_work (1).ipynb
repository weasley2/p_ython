```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# python program where loading data into pandas Dataframe
df = pd.read_csv('DataScienceSalaries.csv')
df
```

```python
# Removing unnecessary columns like:(salary,salary_currency)

removecolumn=['salary_currency', 'salary']
df=df.drop(columns=removecolumn)
df
```

```python
# check missing value details
missingValue = df.isna().sum()
missingValue
```

```python
# null value chaina but vako vaye 
# this is how you will drop it
```

```python
# removing the NaN missing values from updated dataframe

df=df.dropna()
df
```

```python
#program to check duplicates value in the dataframe

df[df.duplicated(keep=False) == True]
df
```

```python
#a python program to see the unique values from all the columns in the dataframe

test=[df[arr].unique() for arr in df.columns]
test
```

```python
# Renaming the columns using expeirence levels

workLevels = {
    "SE": "Senior Level/Expert",
    "MI": "Medium Level/Intermediate",
    "EN": "Entry Level",
    "EX": "Executive Level"
}

df['experience_level'] = df['experience_level'].map(workLevels)
df

# Renaming other columns for data consistency
employmentTypes = {
    "ML Engineer": "Machine Learning Engineer",
    "MLOps Engineer": "Machine Learning Ops Engineer" # This didn't need to be done since ML Ops Enginner was unique (Makes to look nicer)
}

df['job_title'] = df['job_title'].map(employmentTypes).fillna(df['job_title'])
df
```

```python
# Matrix -> 3,4 means 3 rows and 4 cols 

# Corr will only work on number cols (salary, work_year etc...) because it compares the relation between two numbers
coloration = {
    "Year": df["work_year"],
    "Salary": df["salary_in_usd"]
}
 
result = pd.DataFrame(coloration)
 
print(result.corr())
```

```python
#program to show summary statistics of mean
dataToCheck = df["salary_in_usd"]
meanofsalary = dataToCheck.mean()
print("Mean " + str(meanofsalary)) 
```

```python
#program to show summary statistics of standard Deviation
standardDeviation = dataToCheck.std()
print("Standard Deviation " + str(standardDeviation))
```

```python
 #program to show summary statistics of skew
skew = dataToCheck.skew()
print("Skew " + str(skew))

```

```python
#program to show summary statistics of kurtosis
kurtosisData = dataToCheck.kurtosis()
print("Kurtosis Data " + str(kurtosisData)) 
```

```python
#program to show summary statistics of sum

sum = dataToCheck.sum()
print("sum "+ str (sum))
```

```python
# Find to 15 jobs
# (the errors = coerce will send any errors in the parsing to be Nan or None
#df['salary_in_usd'] = pd.to_numeric(df['salary_in_usd'], errors='coerce')
#topJobs = df.groupby('job_title')['salary_in_usd'].mean().sort_values(ascending=False).head(15)
topJobs = df['job_title'].value_counts().head(15)
topJobs
plt.figure(figsize=(10, 8))
topJobs.plot(kind='bar')
plt.colorbar = 'maroon'
plt.title('Top 15 Jobs by Count')
plt.xlabel('Job Title')
plt.ylabel('Job Count')
plt.xticks(rotation=45, ha='right') # This to give it a fancy design at the bottom
plt.tight_layout()
plt.savefig('test')
```

```python
# Highest Salaries
#print(df['salary'])
topJobs = df.groupby('experience_level')['salary_in_usd'].max().sort_values(ascending=False)
plt.figure(figsize=(10, 8))
topJobs.plot(kind='bar')
plt.colorbar = 'maroon'
plt.title('Highest Salary by experience')
plt.xlabel('Job Experience')
plt.ylabel('Max Salary')
plt.xticks(rotation=45, ha='right') # This to give it a fancy design at the bottom
plt.tight_layout()
plt.show() 
```

```python
#finding  out salaries based on experience level. Illustrating it through bar graph.
average_salary = df.groupby('experience_level')['salary_in_usd'].mean()
plt.figure(figsize=(10, 8))
average_salary.plot(kind='bar', color='skyblue')
plt.xlabel('Experience Level')
plt.ylabel('Salary (USD)')
plt.title('')
plt.show() 
```

```python
#program to show histogram and box plot of any chosen different variables,proper labels in the graph.

plt.figure(figsize=(10, 5))
plt.hist(df['salary_in_usd'], bins=20, color='blue', alpha=0.7)
plt.title('Distribution of Salaries in USD')
plt.xlabel('Salary in USD')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

top_3_experience_levels = df['experience_level'].value_counts().head(3).index

filtered_df = df[df['experience_level'].isin(top_3_experience_levels)]
plt.figure(figsize=(15, 5))
filtered_df.boxplot(column='salary_in_usd', by='experience_level')
plt.title('Salary Distribution by Experience Level')
plt.suptitle('')
plt.xlabel('Experience Level')
plt.ylabel('Salary in USD')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()


```
